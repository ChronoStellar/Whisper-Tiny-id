## Implementing model with pipeline to gradio

### Whisper Tiny

```python
from transformers import pipeline
import gradio as gr

# Load the Whisper-tiny Indonesian model
pipe = pipeline(model="openai/whisper-tiny")

def transcribe(audio):
    text = pipe(audio)["text"]
    return text

# Create the Gradio interface
iface = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Fined Tuned Whisper Tiny Indonesian",
    description="Transcribe Indonesian audio using Whisper-tiny model.",
)

# Launch the Gradio app
iface.launch()
```

### Whisper Tiny Fine Tuned

```python
from transformers import pipeline
import gradio as gr

# Load the Whisper-tiny Indonesian model
pipe1 = pipeline(model="ChronoStellar/whisper-tiny-id")

def transcribe(audio):
    text = pipe1(audio)["text"]
    return text

# Create the Gradio interface
iface = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Fined Tuned Whisper Tiny Indonesian",
    description="Transcribe Indonesian audio using Whisper-tiny model.",
)

# Launch the Gradio app
iface.launch()
```

```python
true_text = 'Pemirsa di tengah derasnya desaikan publik atas pengungkapan kasus kematian Fina dan Eki.  Publik dikejutkan dengan munculnya tangkapan layar rekaman CCTV yang diduga berkaitan dengan kasus pembunuhan Fina di tahun.'
tiny_text = 'Kaya pemerisah ditengah drasnya, desakan publik atau sepunggu kapan kasus kematian fina dan eki publik di kejutkan dengan munculnya tankap ala ya rekaman sejiatif yang diduka, berkaitan dengan kasus pembunuhan fina di tahun.'
tuned_text = 'Perapema-mirsat ditengah dirasnya, desakan publik atas pemungkapan kasus kematian Finan dan Eki publik dikecutkan dengan munculnya tanpa layar rekaman sisi Vygyang di Duga, perkaitan dengan kasus pembunuhan Finan di.'
```

```python
!pip install jiwer
```

```python
from jiwer import wer

error1 = wer(true_text, tiny_text)
error2 = wer(true_text, tuned_text)
```

```python
print(f'Tiny WER: {error1}')
print(f'Tuned WER: {error2}')
```
